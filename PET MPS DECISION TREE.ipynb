```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier, export_graphviz                 
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
import graphviz
import matplotlib.pylab as plt

petdata = pd.read_excel ('pet.xlsx')
X = petdata.drop(columns=['PrPET-RE'])
y= petdata['PrPET-RE'].astype(str)
model = DecisionTreeClassifier().fit(X, y)
pipe = Pipeline ([("scale", StandardScaler()),
                  ("model",DecisionTreeClassifier(random_state=0))])

pipe.fit(X, y)
pred = pipe.predict(X)

# Export the decision tree as a .dot file
pettree = export_graphviz(model, out_file='petmp.dot', 
                feature_names=['CIPConc','Temperature','pH','Ionic strength' ],
                class_names=sorted(y.unique()), label='all', 
                filled=True, rounded=True)
```
